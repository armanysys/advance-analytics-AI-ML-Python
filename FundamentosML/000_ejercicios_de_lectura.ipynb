## Introduccion Python

```python
# Definimos dos variables numéricas
numero1 = 8
numero2 = 3

# Realizamos una operación de suma
resultado = numero1 + numero2

# Mostramos el resultado
print("El resultado de la suma es:", resultado)
```

### Control de flujo

```python
# Lista de ventas por vendedor
ventas = [800, 1200, 950, 1100, 500]
meta = 1000  # Meta de ventas

# Recorremos la lista de ventas
for venta in ventas:
    if venta >= meta:
        print(f"Venta de {venta}: Meta alcanzada, comisión aplicada.")
    else:
        print(f"Venta de {venta}: Meta no alcanzada, no hay comisión.")
```

### Estructura de Datos

```python
# Lista de nombres
nombres = ["Carlos", "Ana", "Beatriz", "Laura"]

# Diccionario con las edades
edades = {
    "Carlos": 35,
    "Ana": 28,
    "Beatriz": 42,
    "Laura": 30
}

# Mostrar los nombres y sus respectivas edades
for nombre in nombres:
    if nombre in edades:
        print(f"{nombre} tiene {edades[nombre]} años.")
```

### Funciones y Modularización

```python
# Definir funciones para cálculos matemáticos
def sumar(a, b):
    return a + b

def restar(a, b):
    return a - b

def multiplicar(a, b):
    return a * b

def dividir(a, b):
    if b != 0:
        return a / b
    else:
        return "No se puede dividir por cero"

# Usar las funciones
num1 = 10
num2 = 5

print(f"Suma: {sumar(num1, num2)}")
print(f"Resta: {restar(num1, num2)}")
print(f"Multiplicación: {multiplicar(num1, num2)}")
print(f"División: {dividir(num1, num2)}")
```

### Manejo de errores

```python
def dividir_numeros():
    try:
        # Pedir valores al usuario
        a = float(input("Introduce el primer número: "))
        b = float(input("Introduce el segundo número: "))
        
        # Intentar dividir los números
        resultado = a / b
        print(f"El resultado es: {resultado}")
    
    except ZeroDivisionError:
        print("Error: No se puede dividir por cero.")
    
    except ValueError:
        print("Error: Por favor, introduce un valor numérico válido.")


# Llamar a la función
dividir_numeros()
```

### Pandas - Manipulación y análisis de datos

```python
import pandas as pd

# Crear un DataFrame con 5 personas
data = {'Nombre': ['Ana', 'Juan', 'Pedro', 'María', 'Luis'],
        'Edad': [23, 28, 22, 30, 35],
        'Ciudad': ['Madrid', 'Barcelona', 'Valencia', 'Sevilla', 'Bilbao']}

df = pd.DataFrame(data)

# Filtrar personas mayores de 25 años
df_mayores_25 = df[df['Edad'] > 25]

# Agregar una columna de mayor de 25 años
df['Mayor de 25'] = df['Edad'].apply(lambda x: 'Sí' if x > 25 else 'No')

# Eliminar la columna 'Ciudad'
df_sin_ciudad = df.drop(columns=['Ciudad'])

# Mostrar el resultado
print(df_sin_ciudad)
```

### NumPy - Operación numérica

```python
import numpy as np

# Crear una matriz 4x4 con valores aleatorios entre 1 y 100
matriz_aleatoria = np.random.randint(1, 101, size=(4, 4))

# Sumar filas y columnas
suma_filas = np.sum(matriz_aleatoria, axis=1)
suma_columnas = np.sum(matriz_aleatoria, axis=0)

# Encontrar máximo y mínimo en cada fila
maximos_filas = np.max(matriz_aleatoria, axis=1)
minimos_filas = np.min(matriz_aleatoria, axis=1)

# Media de la matriz
media_matriz = np.mean(matriz_aleatoria)

# Restar la media a cada elemento de la matriz
matriz_normalizada = matriz_aleatoria - media_matriz

print("Matriz Aleatoria:\n", matriz_aleatoria)
print("Suma por filas:", suma_filas)
print("Suma por columnas:", suma_columnas)
print("Máximos por fila:", maximos_filas)
print("Mínimos por fila:", minimos_filas)
print("Matriz Normalizada:\n", matriz_normalizada)
```

## Matplotlib & Seaborn – Basicos sobre la Visualización de Datos

### 1 Introducción a la Creación de Gráficos Básicos con Matplotlib

Matplotlib es una de las librerías más populares para la creación de gráficos en Python. Permite la visualización de datos en distintos formatos de gráfico, como líneas, barras y dispersión.

Ejemplo de Código:

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos simulados de eficiencia de una planta de producción
dias = np.arange(1, 11)  # Días del 1 al 10
eficiencia = [80, 82, 85, 83, 88, 86, 89, 90, 92, 91]  # Eficiencia porcentual diaria

# Crear un gráfico de líneas básico
plt.plot(dias, eficiencia)
plt.title('Eficiencia Operativa de la Planta de Producción')
plt.xlabel('Día')
plt.ylabel('Eficiencia (%)')
plt.show()
```

### 2. Crear Gráficos Simples

Matplotlib permite la creación de varios tipos de gráficos. A continuación, se presentan ejemplos para gráficos de líneas, barras y dispersión.

- Gráfico de Líneas: Ideal para mostrar el rendimiento de la planta a lo largo del tiempo.

```python
plt.plot(dias, eficiencia, color='green', marker='o')
plt.title('Eficiencia Operativa a lo Largo del Tiempo')
plt.xlabel('Día')
plt.ylabel('Eficiencia (%)')
plt.grid(True)
plt.show()
```

- Gráfico de Barras: Útil para comparar métricas de distintos periodos o áreas de la planta.

```python
# Crear un gráfico de barras
plt.bar(dias, eficiencia, color='blue')
plt.title('Eficiencia Operativa por Día')
plt.xlabel('Día')
plt.ylabel('Eficiencia (%)')
plt.show()
```

- Gráfico de Dispersión: Se usa para mostrar relaciones entre dos variables, como la eficiencia y el uso de recursos.

```python
# Simulación de datos de uso de recursos
recursos_usados = [50, 55, 60, 58, 65, 63, 67, 68, 70, 72]

# Crear un gráfico de dispersión
plt.scatter(recursos_usados, eficiencia, color='red')
plt.title('Relación entre el Uso de Recursos y la Eficiencia')
plt.xlabel('Uso de Recursos (%)')
plt.ylabel('Eficiencia (%)')
plt.show()
```

### 3. Personalización de Gráficos

Es importante hacer que los gráficos sean más claros y personalizados. Podemos añadir títulos, etiquetas a los ejes y leyendas para facilitar la interpretación.

Ejemplo de Código:

```python
plt.plot(dias, eficiencia, color='purple', linestyle='--', marker='s', label='Eficiencia')
plt.title('Eficiencia Operativa de la Planta')
plt.xlabel('Día')
plt.ylabel('Eficiencia (%)')
plt.legend(loc='upper left')  # Añadir leyenda
plt.grid(True)
plt.show()
```

### 4. Guardar Gráficos

Para guardar los gráficos generados como imágenes, podemos usar el método savefig().

Ejemplo de Código:

```python
plt.plot(dias, eficiencia, color='blue', marker='o')
plt.title('Eficiencia Operativa a lo Largo del Tiempo')
plt.xlabel('Día')
plt.ylabel('Eficiencia (%)')

# Guardar el gráfico como archivo PNG
plt.savefig('img/eficiencia_operativa.png')
plt.show()
```

### 5. Gráficos de Líneas para Visualizar la Eficiencia a lo Largo del Tiempo

El gráfico de líneas es una excelente opción para visualizar cómo varía la eficiencia operativa a lo largo del tiempo. En este caso, podemos comparar la eficiencia de diferentes áreas de producción.

Ejemplo de Código:

```python
# Simulación de datos de eficiencia de dos áreas de la planta
eficiencia_area_1 = [80, 82, 84, 85, 87, 85, 88, 89, 90, 92]
eficiencia_area_2 = [75, 78, 80, 83, 82, 85, 86, 87, 89, 91]

# Gráficos de líneas para comparar áreas
plt.plot(dias, eficiencia_area_1, label='Área 1', color='green', marker='o')
plt.plot(dias, eficiencia_area_2, label='Área 2', color='red', marker='s')
plt.title('Comparación de Eficiencia entre Áreas de Producción')
plt.xlabel('Día')
plt.ylabel('Eficiencia (%)')
plt.legend()
plt.grid(True)
plt.show()
```

### 6. Mapas de Calor de Métricas Operativas con Seaborn

Seaborn facilita la creación de mapas de calor, que son ideales para visualizar métricas operativas en función de diferentes variables.

Ejemplo de Código:

```python
import seaborn as sns

# Simulación de datos de eficiencia y uso de recursos en varias áreas de la planta
datos = np.random.rand(10, 5) * 100  # Datos simulados de eficiencia en 5 áreas durante 10 días
areas = ['Área 1', 'Área 2', 'Área 3', 'Área 4', 'Área 5']

# Crear un mapa de calor
sns.heatmap(datos, annot=True, cmap='coolwarm', xticklabels=areas, yticklabels=dias)
plt.title('Mapa de Calor de la Eficiencia en Áreas de Producción')
plt.xlabel('Área de Producción')
plt.ylabel('Día')
plt.show()
```

### 7. Gráficos de Barras y Distribución de Métricas Operativas

Los gráficos de barras y distribuciones ayudan a comparar el uso de recursos o la eficiencia entre distintas áreas o periodos de tiempo.

- Gráfico de Barras:

```python
# Simulación de datos de uso de recursos
recursos_usados = [80, 85, 75, 90, 88]

# Crear un gráfico de barras
plt.bar(areas, recursos_usados, color='cyan')
plt.title('Uso de Recursos por Área de Producción')
plt.xlabel('Área')
plt.ylabel('Uso de Recursos (%)')
plt.show()
```

- Distribución con Seaborn:

```python
# Simulación de datos de eficiencia de una planta
eficiencia_total = np.random.normal(loc=85, scale=5, size=100)

# Crear gráfico de distribución con Seaborn
sns.histplot(eficiencia_total, kde=True, color='purple')
plt.title('Distribución de la Eficiencia Operativa')
plt.xlabel('Eficiencia (%)')
plt.ylabel('Frecuencia')
plt.show()
```
