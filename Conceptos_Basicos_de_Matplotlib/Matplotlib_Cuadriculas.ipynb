```python
# Usando `ìnline` no sea necesario escribir código adicional `plt.show()` para mostrar los gráficos
%matplotlib inline

import pandas as pd
import numpy as np

# `plt` es un alias para el módulo `matplotlib.pyplot`
import matplotlib.pyplot as plt

# importa libreria seaborn (contenedor de matplotlib)
import seaborn as sns
```

### Carga los datos

```python
# Cargar datos de préstamos de automóviles en un marco de datos de pandas desde un archivo csv
filename = '/content/table_i702t60.csv'
df = pd.read_csv(filename)
```

```python
# Ver las primeras 5 filas del marco de datos
df.head()
```

```python
# Comprobando que no tenemos `NaN` en nuestro DataFrame
# No es fácil trazar directamente datos que contienen `NaN`
df.info()
```

```python
# Para este cuaderno, graficaremos interes_pagado, capital_pagado y mes en un gráfico
# Si bien podríamos graficar directamente a través de pandas, por ahora lo haremos a través de matplotlib.
mes_numero = df.loc[:, 'mes'].values
interes_pagado = df.loc[:, 'interes_pagado'].values
capital_pagado = df.loc[:, 'capital_pagado'].values
```

```python
mes_numero
```

```python
# The values attribute converts a column of values into a numpy array
type(mes_numero)
```

## Grids

```python
plt.plot(mes_numero, interes_pagado, c= 'k')
plt.plot(mes_numero, capital_pagado, c = 'b')
```

### Graficos al estilo MATLAB

```python
plt.plot(mes_numero, interes_pagado, c= 'k')
plt.plot(mes_numero, capital_pagado, c = 'b')
plt.grid()
```

```python
# solo líneas de cuadrícula horizontales
plt.plot(mes_numero, interes_pagado, c= 'k')
plt.plot(mes_numero, capital_pagado, c = 'b')
plt.grid(axis = 'y')
```

```python
# solo líneas de cuadrícula verticales
plt.plot(mes_numero, interes_pagado, c= 'k')
plt.plot(mes_numero, capital_pagado, c = 'b')
plt.grid(axis = 'x')
```

```python
# cambiar el color de las líneas de la cuadrícula, la transparencia y el estilo de línea
plt.plot(mes_numero, interes_pagado, c= 'k')
plt.plot(mes_numero, capital_pagado, c = 'b')
plt.grid(c = 'g',
         alpha = .9,
         linestyle = '-')
```

### Orientado a Objetos

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1);
axes.plot(mes_numero, interes_pagado, c= 'k');
axes.plot(mes_numero, capital_pagado, c = 'b');
axes.grid()
```

```python
# solo líneas de cuadrícula horizontales
fig, axes = plt.subplots(nrows = 1, ncols = 1);
axes.plot(mes_numero, interes_pagado, c= 'k');
axes.plot(mes_numero, capital_pagado, c = 'b');
axes.grid(axis = 'y')
```

```python
# solo líneas de cuadrícula verticales
fig, axes = plt.subplots(nrows = 1, ncols = 1);
axes.plot(mes_numero, interes_pagado, c= 'k');
axes.plot(mes_numero, capital_pagado, c = 'b');
axes.grid(axis = 'x')
```

```python
# cambiar el color de las líneas de la cuadrícula, la transparencia y el estilo de línea
fig, axes = plt.subplots(nrows = 1, ncols = 1);
axes.plot(mes_numero, interes_pagado, c= 'k');
axes.plot(mes_numero, capital_pagado, c = 'b');
axes.grid(c = 'g',
          alpha = .9,
          linestyle = '-')
```
