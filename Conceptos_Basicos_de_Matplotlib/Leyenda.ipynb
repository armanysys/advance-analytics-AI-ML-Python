## Leyendas

El parámetro `loc` (ubicación de la leyenda) acepta cadenas, enteros y tuplas

string | int
--- | ---
'best' | 0
'upper right' | 1
'upper left' | 2
'lower left' |3
'lower right' | 4
'right' | 5
'center left' | 6
'center right' | 7
'lower center' | 8
'upper center' | 9
'center' | 10

El parámetro acepta una tupla de 2 elementos ``x, y`` donde (0, 0) es la esquina inferior izquierda de la leyenda en coordenadas de ejes.

### MATLAB-style

```python
# Obviamente la leyenda no está en una ubicación ideal.
plt.plot(mes_numero, interes_pagado, c= 'k', label = 'Interes')
plt.plot(mes_numero, capital_pagado, c = 'b', label = 'Capital')
plt.legend()
```

```python
# Al menos la leyenda no se superpone con el gráfico.
plt.plot(mes_numero, interes_pagado, c= 'k', label = 'Interes')
plt.plot(mes_numero, capital_pagado, c = 'b', label = 'Capital')
plt.legend(loc="center right")
```

```python
# Puedes mover la leyenda fuera del área de trazado.
# Al menos la leyenda no se superpone con el gráfico.
plt.plot(mes_numero, interes_pagado, c= 'k', label = 'Interes')
plt.plot(mes_numero, capital_pagado, c = 'b', label = 'Capital')
plt.legend(loc=(1.02,0))
```

### Object-oriented

```python
# Obviamente la leyenda no está en una ubicación ideal.
fig, axes = plt.subplots(nrows = 1, ncols = 1);
axes.plot(mes_numero, interes_pagado, c= 'k', label = 'Interes');
axes.plot(mes_numero, capital_pagado, c = 'b', label = 'Capital');
axes.legend()
```
