```python
# Usando `ìnline` no sea necesario escribir código adicional `plt.show()` para mostrar los gráficos
%matplotlib inline

import pandas as pd
import numpy as np

# `plt` es un alias para el módulo `matplotlib.pyplot`
import matplotlib.pyplot as plt

# importa libreria seaborn (contenedor de matplotlib)
import seaborn as sns
```

### Cargando Datos para nuestros graficos

```python
# Cargue datos de préstamos de automóviles en un marco de datos de pandas desde un archivo csv
filename = '/content/table_i702t60.csv'
df = pd.read_csv(filename)
```

```python
# Ver las primeras 5 filas del marco de datos
df.head()
```

```python
# Comprobando que no tenemos `NaN` en nuestro DataFrame
# No es fácil trazar directamente datos que contienen `NaN`
df.info()
```

```python
# Para este cuaderno, graficaremos los intereses pagados, el capital pagado y el mes en un solo gráfico.
# Si bien podríamos graficar directamente a través de Pandas, por ahora lo haremos a través de Matplotlib.
mes_numero = df.loc[:, 'mes'].values
interes_pagado = df.loc[:, 'interes_pagado'].values
capital_pagado = df.loc[:, 'capital_pagado'].values
```

## Establecer títulos, etiquetas y límites

### Estilo MATLAB

```python
plt.plot(mes_numero, interes_pagado, c= 'k')
plt.plot(mes_numero, capital_pagado, c = 'b')
```

#### Establece xlim y ylim

```python
# This isn't the most practical use of changing ylim
plt.plot(mes_numero, interes_pagado, c= 'k')
plt.plot(mes_numero, capital_pagado, c = 'b')
plt.xlim(left=1,right=70)
plt.ylim(bottom=0,top=1000)
```

#### Estableciendo xlabel y ylabel

```python
# Obviously this isnt the most practical use of changing xlim and ylim
plt.plot(mes_numero, interes_pagado, c= 'k')
plt.plot(mes_numero, capital_pagado, c = 'b')
plt.xlabel('Mes')
plt.ylabel('Pesos')
```

#### Estableciendo Titulo

```python
plt.plot(mes_numero, interes_pagado, c= 'k')
plt.plot(mes_numero, capital_pagado, c = 'b')
plt.xlabel('Mes')
plt.ylabel('Pesos')
plt.title('Intereses y Capital pagado por Mes')
```

#### Cambiando el tamaño de las letras

```python
plt.plot(mes_numero, interes_pagado, c= 'k')
plt.plot(mes_numero, capital_pagado, c = 'b')
plt.xlabel('Mes', fontsize = 15)
plt.ylabel('Pesos', fontsize = 15)
plt.title('Intereses y Capital pagado por Mes', fontsize = 15)
```

```python
# Cambiando el tamaño de mas marcas
plt.plot(mes_numero, interes_pagado, c= 'k')
plt.plot(mes_numero, capital_pagado, c = 'b')
plt.xlabel('Mes', fontsize = 15)
plt.ylabel('Pesos', fontsize = 15)
plt.title('Intereses y Capital pagado por Mes', fontsize = 15)
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
```

### Codigo Orientado a Objetos

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(mes_numero, interes_pagado, c= 'k');
axes.plot(mes_numero, capital_pagado, c = 'b');
```

#### Estableciendo los limites xlim y ylim

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(mes_numero, interes_pagado, c= 'k');
axes.plot(mes_numero, capital_pagado, c = 'b');
axes.set_xlim(left =1 , right = 70)
axes.set_ylim(bottom = 0, top = 1000)
```

#### Estableciendo xlabel and ylabel

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1)
axes.plot(mes_numero, interes_pagado, c= 'k');
axes.plot(mes_numero, capital_pagado, c = 'b');
axes.set_xlabel('Mes')
axes.set_ylabel('Pesos')
```

#### Estableciendo el Titulo

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1);
axes.plot(mes_numero, interes_pagado, c= 'k');
axes.plot(mes_numero, capital_pagado, c = 'b');
axes.set_xlabel('Mes');
axes.set_ylabel('Pesos');
axes.set_title('Intereses y Capital pagado por Mes');
```

#### Cambiando el tamaño de la tipografia

```python
fig, axes = plt.subplots(nrows = 1, ncols = 1);
axes.plot(mes_numero, interes_pagado, c= 'k');
axes.plot(mes_numero, capital_pagado, c = 'b');
axes.set_xlabel('Mes', fontsize = 22);
axes.set_ylabel('Pesos', fontsize = 22);
axes.set_title('Intereses y Capital pagado por Mes', fontsize = 22);
```

```python
# Cambiando el tamaño de las marcas
fig, axes = plt.subplots(nrows = 1, ncols = 1);
axes.plot(mes_numero, interes_pagado, c= 'k');
axes.plot(mes_numero, capital_pagado, c = 'b');
axes.set_xlabel('Mes', fontsize = 22);
axes.set_ylabel('Pesos', fontsize = 22);
axes.set_title('Intereses y Capital pagado por Mes', fontsize = 22);
axes.tick_params(axis = 'x', labelsize = 20)
axes.tick_params(axis = 'y', labelsize = 20)
```
