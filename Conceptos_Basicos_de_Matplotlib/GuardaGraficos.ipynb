```python
# Usando `ìnline` no sea necesario escribir código adicional `plt.show()` para mostrar los gráficos
%matplotlib inline

import pandas as pd
import numpy as np

# `plt` es un alias para el módulo `matplotlib.pyplot`
import matplotlib.pyplot as plt

# importa libreria seaborn (contenedor de matplotlib)
import seaborn as sns
```

### Cargando Datos para nuestros graficos

```python
# Cargue datos de préstamos de automóviles en un marco de datos de pandas desde un archivo csv
filename = '/content/table_i702t60.csv'
df = pd.read_csv(filename)
```

```python
# Ver las primeras 5 filas del marco de datos
df.head()
```

```python
# Comprobando que no tenemos `NaN` en nuestro DataFrame
# No es fácil trazar directamente datos que contienen `NaN`
df.info()
```

```python
# Para este cuaderno, graficaremos los intereses pagados, el capital pagado y el mes en un solo gráfico.
# Si bien podríamos graficar directamente a través de Pandas, por ahora lo haremos a través de Matplotlib.
mes_numero = df.loc[:, 'mes'].values
interes_pagado = df.loc[:, 'interes_pagado'].values
capital_pagado = df.loc[:, 'capital_pagado'].values
```

## Guardar gráficos en archivos
Guardar las visualizaciones fuera de su cuaderno Jupyter es importante, ya que le permite mostrarlas a otros. Igualmente importante es verificar la visualización guardada, ya que siempre existe la posibilidad de que el gráfico no se vea igual en el cuaderno que en el archivo de imagen.

```python
plt.style.use('seaborn-whitegrid')
```

### MATLAB-style

```python
# an image may good in the notebook, but it may not save that way
plt.figure(figsize=(10, 5))
plt.plot(mes_numero, capital_pagado, c = 'b', label = 'Capital')
plt.plot(mes_numero, interes_pagado, c= 'k', label = 'Interes')
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlim(left =1 , right = 61)
plt.ylim(bottom = 0, top = 700)
plt.xlabel('Mes', fontsize = 22);
plt.ylabel('Pesos', fontsize = 22);
plt.title('Interes y Capital Pagado por Mes', fontsize = 24)
plt.legend(loc=(1.02,0), borderaxespad=0, fontsize = 20)

plt.savefig('mslegendcutoff.png', dpi = 300)
```

```python
# tight_layout()
# automatically adjusts subplot params so that the subplot(s) fits in to the figure area
plt.figure(figsize=(10, 5))
plt.plot(mes_numero, capital_pagado, c = 'b', label = 'Capital')
plt.plot(mes_numero, interes_pagado, c= 'k', label = 'Interes')
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.xlim(left =1 , right = 61)
plt.ylim(bottom = 0, top = 700)
plt.xlabel('Mes', fontsize = 22);
plt.ylabel('Pesos', fontsize = 22);
plt.title('Interes y Capital Pagado por Mes', fontsize = 24)
plt.legend(loc=(1.02,0), borderaxespad=0, fontsize = 20)

plt.tight_layout()
plt.savefig('mslegend.png', dpi = 300)
```

### Object-oriented

```python
# an image may good in the notebook, but it may not save that way
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize=(10, 5) )
plt.plot(mes_numero, capital_pagado, c = 'b', label = 'Capital')
plt.plot(mes_numero, interes_pagado, c= 'k', label = 'Interes')
axes.tick_params(axis = 'x', labelsize = 20)
axes.tick_params(axis = 'y', labelsize = 20)
axes.set_xlim(left =1 , right = 61)
axes.set_ylim(bottom = 0, top = 700)
axes.set_xlabel('Mes', fontsize = 22);
axes.set_ylabel('Pesos', fontsize = 22);
axes.set_title('Interes y Capital Pagado por Mes', fontsize = 24)
axes.legend(loc=(1.02,0), borderaxespad=0, fontsize = 20)

fig.savefig('objectlegendcutoff.png', dpi = 300)
```

```python
# tight_layout()
# automatically adjusts subplot params so that the subplot(s) fits in to the figure area
fig, axes = plt.subplots(nrows = 1, ncols = 1, figsize=(10, 5) )
axes.plot(mes_numero, capital_pagado, c = 'b', label = 'Capital')
axes.plot(mes_numero, interes_pagado, c= 'k', label = 'Interes')
axes.tick_params(axis = 'x', labelsize = 20)
axes.tick_params(axis = 'y', labelsize = 20)
axes.set_xlim(left =1 , right = 61)
axes.set_ylim(bottom = 0, top = 700)
axes.set_xlabel('Mes', fontsize = 22);
axes.set_ylabel('Pesos', fontsize = 22);
axes.set_title('Interes y Capital Pagado por Mes', fontsize = 24)
axes.legend(loc=(1.02,0), borderaxespad=0, fontsize = 20)

fig.tight_layout()
fig.savefig('objectlegend.png', dpi = 300)
```
